# CS6220 HW1 Experiments Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## WordCount on SMS Spam Collection

```python
def plot_bar(y, ylabel, x, xlabel, title, data):
    sns.catplot(y=y, x=x, kind='bar', data=data)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

### 1. Default Setting Execution Time

```python
data = {'map_occupied': [6708, 34964], 'map': [1677, 8741], 'reduce_occupied': [13240, 13736], 'reduce': [1655, 1717], 'type': ['small', 'large']}
data = pd.DataFrame(data=data)
```

```python
plot_bar('map_occupied', 'Time spent (ms)', 'type', 'Dataset', 'Time spent by maps in occupied slots', data)
```

```python
plot_bar('reduce_occupied', 'Time spent (ms)', 'type', 'Dataset', 'Time spent by reduces in occupied slots', data)
```

```python
plot_bar('map', 'Time spent (ms)', 'type', 'Dataset', 'Time spent by all maps', data)
```

```python
plot_bar('reduce', 'Time spent (ms)', 'type', 'Dataset', 'Time spent by all reduces', data)
```

### 2. Relationship between Input Split Size and Execution Time

#### Small Dataset

```python
data = {'map_occupied': [13772, 20284], 'map': [3443, 5071], 'split_size': [100000, 500000]}
data = pd.DataFrame(data=data)
```

```python
plot_bar('map_occupied', 'Time spent (ms)', 'split_size', 'Split Size', 'Execution time v Split Size (Small Dataset)', data)
```

#### Large Dataset

```python
data = {'map_occupied': [42484, 41644], 'map': [10621, 10411], 'split_size': [100000, 500000]}
data = pd.DataFrame(data=data)
```

```python
plot_bar('map_occupied', 'Time spent (ms)', 'split_size', 'Split Size', 'Execution time v Split Size (Large Dataset)', data)
```
